### **B) Language Understanding (LUIS)**

Language Understanding (LUIS) is a cloud-based API service that uses machine learning to the natural language text of a user's conversations to predict their overall meaning. A client application for LUIS is a conversation application that communicates with a user in natural language to accomplish a task. 

The bot created before contains the entire folders JSON that allow us to create and edit the intents and entities. For this, we have 2 choices: 
- Edit the code directly in the file 
- Import the file and edit it in LUIS 

For our project we will import the file to manage it in LUIS with the additional functionality that LUIS provides us. To do this, we need to copy the JSON file and extract it locally on our machine. Once this is done we will have to import it into LUIS and make a link between the webapp bot and LUIS. We will do these different steps in the following section **LUIS**.

The LUIS service enables the user to build and customize on his own the language models, allowing any application or bot to understand different commands and act accordingly. In fact, LUIS is able to extract from a sentence (an order, or a question in our case) possible intents, its score (how much they match the sentence), as well as the different entities: 
![alt text](https://github.com/ounon/BSA2020_Team_Google_Project_1/blob/master/Pictures/LUISpic.jpg?raw=true)

####Definition
Here are some basic concepts to help you understand the use of this tool.

-**Intent**
The intents define the intentions of the users. What the user actually wants to do and meant with the sentence/question to the chatbot. Note that, if what the user says does not match exactly (spelling mistake, slightly different word) the Intent can still be identified as far as possible.

-**Entities**
Entities correspond to parameters of what the user says. These parameters can be a list, they can also be mandatory. You just have to check the right options. 
They are important words in a sentence/ request that will act like parameters within the intent function. 

For example: <br>
Question by the user: "I would like to order a pizza Margherita". <br>
For a chatbot of a Pizza restaurant/ delivery service the **intent** would be ORDER PIZZA and the **entity** MARGHERITA.

####**2. Code implementation**








```
var request = require('request');
var restify = require('restify');
var builder = require('botbuilder');


// Connector
var connectorAppId = process.env.MicrosoftAppId;
var connectorAppPassword = process.env.MicrosoftAppPassword;

var inMemoryStorage = new builder.MemoryBotStorage();

var openWeatherMapKey = YOUR_OPENWETHERMAP_KEY;

// LUIS model
var luisModelUrl = 'https://westus.api.cognitive.microsoft.com/luis/v2.0/apps/YOUR_APPLICATION_ID?subscription-key=YOUR_APPLICATION_KEY';


// Create Restify Server
var server = restify.createServer();
server.listen(process.env.port || process.env.PORT || 3978, function () {
   console.log('%s listening to %s', server.name, server.url); 
});
  
// Create connector and bot
var connector = new builder.ChatConnector({
    appId: connectorAppId,
    appPassword: connectorAppPassword
});

var bot = new builder.UniversalBot(connector).set('storage', inMemoryStorage);;
server.post('/api/messages', connector.listen());

// Create LUIS recognizer 
var recognizer = new builder.LuisRecognizer(luisModelUrl);
var dialog = new builder.IntentDialog({ recognizers: [recognizer] });
bot.dialog('/', dialog);


// Adding dialog
dialog.matches('GetWeather', [
    function (session, args, next) {
        var city = builder.EntityRecognizer.findEntity(args.entities, 'city');
        session.send("looks like you want to know the weather")
        
        if (!city) {
            builder.Prompts.text(session, 'Which city ?');  
        } else {
            next({ response: city.entity });
        }
    },
    
    function (session, results) {
        if (results.response){
            openweathermap(results.response, function(success, previsions) {
                if (!success) return session.send('An error has occurred');
                                    
                session.send('It is ' + previsions.temperature + 'Â°C in '
                             + previsions.city + ' with ' + previsions.description;);
            });
        }
       
    }
]);

dialog.onDefault(function (session) {
    session.send('I do not understand what you mean');
});


var openweathermap = function(city, callback){
	var url = 'http://api.openweathermap.org/data/2.5/weather?q=' + city + '&lang=en&units=metric&appid=' + openWeatherMapKey;
	
	request(url, function(err, response, body){
		try{		
			var result = JSON.parse(body);
			
			if (result.cod != 200) {
				callback(false);
			} else {
				var previsions = {
					temperature : Math.round(result.main.temp),
					city : result.name,
                    description: result.weather[0].description
				};
						
				callback(true, previsions);
			}
		} catch(e) {
			callback(false); 
		}
	});
}
```

We need the modules botbuilder to create the bot, restify to create a server and request to call the API OpenWeatherMap that we use to get the information about the weather in a specific city. 

To use the openweathermap api, it is necessary to create an account on the site [https://openweathermap.org/](https://openweathermap.org/) so you will get a key that you will use to call the API.

The variable *luisModelUrl* stores the URL of our model luis previously created.

The *bot* variable is used to create our bot.

The variable *recogniser* makes it possible to recognize the itents present in our model.

Once our bot linked to luis we can create the dialogues we want. In our case if in a dialog the bot identifies the intention *GetWeather* it will return **"Look like you want to know the weather"** and if it doesn't detect a city it will ask the user with the question **"Which city?"**. After it is aware of the city it will return the current temperature in the desired city using the openweathermap API. 

However, if it doesn't recognize the GetWeather intention, it will return **"I do not understand what you mean"**.

## **D) QnA Maker and Knowledge Base** - *(Coronavirus case)*

**Question & Answer (QnA)**: QnA Maker is one of the AI services that can help a bot answer users' natural language questions by matching them with the best possible answers from an associated knowledge base.

The simpliest way to feed a ChatBot using Microsoft Azure is through a **QnA knowledge base** commonly known as FAQ *(A file containing a bunch of potential questions that the user may query and their respective questions)*

> In the following [link](https://docs.microsoft.com/en-us/learn/modules/build-a-faq-chat-bot-with-qna-maker-and-azure-bot-service/) is described the steps to create it.

There are two options to populate the knowledge base: <br>(i) Using an existing source either copying the url or uploading a pdf/csv file. <br>(ii) Creating manually the FAQ file in the platform.

###What is QnA knowledge base?
QnA Maker is a cloud-based API service that allows you to implement a question/answer knowledge base and interface as a chat. We need to import our data on the platform in order for the service to interface the chat.
A knowledge base consists of sets of questions and answers provided by URLs and public files. Our bot will answer questions by matching them with the best possible answer from the knowledge base that we created. It is also possible to use the chit-chat file but we did not use it.
